## Step 4
Для кожної з вибірок намалюйте графік на якому будуть зображені точки, що попадають всередину інтервалу [-1,1] та зафарбуйте їх зеленим кольором, що попадають в
інтервали [-3,-1) та (1,3] та зафарбуйте їх синім, а також точки, що виходять за межі інтервалу [-3,3] та зафарбуйте їх червоним. По осі x відкладіть номер елемнта, а по осі
y його значення. Розфарбуйте смуги ( [-1,1], [-3,-1) та (1,3], та все що поза [-3,3]) у блідо зелений, синій та червоний кольори.


```python
def plot_this_strange_stuff(sample, fill=True):
    groups = {} 
    groups['green'] = (sample >= -1) & (sample <= 1)
    groups['blue'] = ((sample >= -3) & (sample <= 3)) & ((sample < -1) | (sample > 1))
    groups['red'] = (~groups['green']) & (~groups['blue'])
    
    for color, ind in groups.items():
        plt.scatter(sample[ind].index, sample[ind], color=color)
        
    if fill:
        plt.fill_between(x=sample.index, y1=-1, y2=1, color='green', alpha=0.1)
        plt.fill_between(x=sample.index, y1=-3, y2=-1, color='blue', alpha=0.1)
        plt.fill_between(x=sample.index, y1=1, y2=3, color='blue', alpha=0.1)
        plt.fill_between(x=sample.index, y1=3, y2=sample.max(), color='red', alpha=0.1)
        plt.fill_between(x=sample.index, y1=sample.min(), y2=-3, color='red', alpha=0.1)
```

```python
plot_this_strange_stuff(df.x)
```

```python
plot_this_strange_stuff(df.z[abs(df.z)<4])
```

Знайдіть вибіркові середні та дисперсії для обох вибірок, та позначте середнє значення жирною зеленою лінією на малюнках. Оцініть ймовірність того що випадкова величина

### Norm

```python
df.x.mean()
```

```python
df.x.var()
```

```python
plot_this_strange_stuff(df.x)
plt.hlines(xmin=df.x.index[0], xmax=df.x.index[-1], y=df.x.mean(), colors='black', linewidth=3)
```

$P(|x-\mathbb{E}x|) > 3\sigma)$

```python
(abs(df.x - df.x.mean()) > 3 * df.x.std(ddof=0)).sum() / df.x.shape[0]
```

### Cauchy

```python
df.z.std(ddof=0)
```

```python
plot_this_strange_stuff(df.z[abs(df.z) < 4])
plt.hlines(xmin=df.z.index[0], xmax=df.z.index[-1], y=df.z.mean(), colors='black', linewidth=3)
```

$P(|x-\mathbb{E}x|) > 3\sigma)$

```python
(abs(df.z - df.z.mean()) > 3 * df.z.std(ddof=0)).sum() / df.z.shape[0]
```
